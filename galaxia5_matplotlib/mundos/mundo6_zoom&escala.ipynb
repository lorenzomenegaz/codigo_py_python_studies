# Mundo 6 - Mudando a escala de X ou Y

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime
from datetime import timedelta
```

```python
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)
```

```python
cotacoes = yf.download(["WEGE3.SA", "PETR4.SA", "VALE3.SA", "LREN3.SA"])['Adj Close']

um_ano_atras = datetime.now() - timedelta(days = 365)

fig, ax = plt.subplots(2, 2, figsize = (20, 7))

coluna = 0

for i in range(2):
    for j in range(2):
        
        ax[i, j].set_xlim(um_ano_atras, datetime.now())
        
        if cotacoes.columns[coluna] == "LREN3.SA":
        
            ax[i, j].set_ylim(20, 35)
        
        ax[i, j].plot(cotacoes.index, cotacoes.iloc[:, coluna])
        ax[i, j].set_title(cotacoes.columns[coluna], color = "w")
        
        coluna = coluna + 1
```

```python
cotacoes = yf.download(["^GSPC"], "1970-01-01")['Adj Close']

fig, ax = plt.subplots(2, 1, figsize = (20, 20))

ax[0].plot(cotacoes.index, cotacoes.values)

ax[1].plot(cotacoes.index, cotacoes.values)
ax[1].set_yscale('log')
```

* Exercício 94: Mande o usuário escolher uma ação. Em seguida, crie 4 gráficos de cotação da ação com os seguintes períodos: 3 anos, 1 ano, 6 meses 1 mês. Ajuste o gráfico tanto no eixo X quanto no eixo Y.
* Exercício 95: Puxe as cotações do ibovespa desde 1994 e trace dois gráficos: um na escala linear e outro na logaritma. 

```python
#gabarito 94

acao_escolhida = str(input("Digite um código de negociação: "))

um_ano_atras = datetime.now() - timedelta(days = 365)
seis_meses = datetime.now() - timedelta(days = 180)
um_mes = datetime.now() - timedelta(days = 30)
tres_anos_atras = datetime.now() - timedelta(days = 1095)

cotacoes = yf.download(acao_escolhida + ".SA")['Adj Close']

fig, ax = plt.subplots(2, 2, figsize = (20, 7))

periodos = [um_mes, seis_meses, um_ano_atras, tres_anos_atras]

for periodo in periodos:
    
    if periodo == um_mes:
        
        cot_periodo = cotacoes[cotacoes.index > periodo]
        ax[0, 0].set_xlim(periodo, datetime.now())
        ax[0, 0].set_ylim(np.min(cot_periodo) - 5, np.max(cot_periodo) + 5)
        ax[0, 0].plot(cot_periodo.index, cot_periodo.values)
        
    elif periodo == seis_meses:
        
        cot_periodo = cotacoes[cotacoes.index > periodo]
        ax[0, 1].set_xlim(periodo, datetime.now())
        ax[0, 1].set_ylim(np.min(cot_periodo) - 5, np.max(cot_periodo) + 5)
        ax[0, 1].plot(cot_periodo.index, cot_periodo.values)
    
    elif periodo == um_ano_atras:
        
        cot_periodo = cotacoes[cotacoes.index > periodo]
        ax[1, 0].set_xlim(periodo, datetime.now())
        ax[1, 0].set_ylim(np.min(cot_periodo) - 5, np.max(cot_periodo) + 5)
        ax[1, 0].plot(cot_periodo.index, cot_periodo.values)
        
    else:
        
        cot_periodo = cotacoes[cotacoes.index > periodo]
        ax[1, 1].set_xlim(periodo, datetime.now())
        ax[1, 1].set_ylim(np.min(cot_periodo) - 5, np.max(cot_periodo) + 5)
        ax[1, 1].plot(cot_periodo.index, cot_periodo.values)
        
        
plt.show()
```
